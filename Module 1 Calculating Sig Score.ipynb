```python
import requests
from bs4 import BeautifulSoup
```

```python
def print_news_links(year, month, day):
    if month == 3:
        day += 43890
    elif month == 4:
        day += 43921
    else:
        day += 43951

    url = f"https://timesofindia.indiatimes.com/2020/{month}/{day}/archivelist/year-{year},month-{month},starttime-{day}.cms"
    page_request = requests.get(url)
    data = page_request.content
    soup = BeautifulSoup(data, "html.parser")

    news_divs = soup.find_all('div', style="font-family:arial ;font-size:12;font-weight:bold; color: #006699")

    news_links = []  # Initialize an empty list to store the URLs
    for div in news_divs:
        links = div.find_all('a')
        for link in links:
            news_links.append(link.get('href'))  # Append each URL to the list

    return news_links

# Example usage:
news_links_list = print_news_links(2020, 4, 5)

den = len(news_links_list)

def search_event_in_news_links(event, news_links_list):
    matching_links = []

    for link in news_links_list:
        if event in link:
            matching_links.append(link)

    return matching_links

# Example usage:
event = "candle"
matching_links = search_event_in_news_links(event, news_links_list)

match_list = []
if matching_links:
    print("Matching links:")
    for link in matching_links:
        print(link)
        match_list.append(link)
else:
    print("No matching links found.")

matching_links = search_event_in_news_links('diya', news_links_list)
num = len(matching_links)


print(num)
print(den)

significance_score = num/den
print(significance_score)
var1 = significance_score
```

```python
import csv
import re

def analyze_event(event, date):
    csv_file_path = f"C:/Users/PIYUSH/Downloads/{date}.csv"

    # Initialize an empty list to store the data
    data_list = []
    # Read data from the CSV file
    with open(csv_file_path, mode='r', newline='', encoding='utf-8') as csv_file:
        reader = csv.DictReader(csv_file)
        for row in reader:
            data_list.append(row)

    event_count = 0
    total_tweets = 0  # Initialize the total tweets counter

    # Print the read data and count total tweets and event occurrences
    for data in data_list:
        if data["content"]:  # Check if content is present
            total_tweets += 1  # Increment total tweets counter
            if re.findall(event, data["content"], flags=re.IGNORECASE):
                event_count += 1

    return event_count, total_tweets

date = input("Enter the date: ")

counter = 0
for i in filtered_keywords:
    event_occurrences, total_tweets_keyword = analyze_event(i, date)
    print(f"Event Occurrences for '{i}':", event_occurrences)
    counter += event_occurrences  # Update counter with event occurrences

print("Total Tweets:", total_tweets_keyword)  # Print total tweets of the last keyword
print("Total Event Occurrences:", counter)
```

```python
import csv
import re

def calculate_emotion(date, emotion, event):
    emotion = emotion.lower()
    csv_file_path = f"C:/Users/PIYUSH/Downloads/{date}_labelled.csv"

    emotion_numbers = []  # Changed from emotion_number to emotion_numbers

    data_list = []
    # Read data from the CSV file
    with open(csv_file_path, mode='r', newline='', encoding='utf-8') as csv_file:
        reader = csv.DictReader(csv_file)
        for row in reader:
            data_list.append(row)

    for data in data_list:
        if re.findall(event, data["content"], flags=re.IGNORECASE):
            emotion_numbers.append(float(data[emotion]))  # Changed to append float value

    return emotion_numbers

date = input("Enter the date: ")
total_emotion_numbers = []  # Changed from total_emotion_number to total_emotion_numbers

for i in filtered_keywords:
    emotion_numbers = calculate_emotion(date, "joy", i)  # Specify the emotion as a string
    print(f"Emotion Numbers for '{i}':", emotion_numbers)
    total_emotion_numbers.extend(emotion_numbers)  # Use extend instead of append

# Calculate the average emotion number
average_emotion_number = sum(total_emotion_numbers) / len(total_emotion_numbers)
print("Average Emotion Number:", average_emotion_number)
```
